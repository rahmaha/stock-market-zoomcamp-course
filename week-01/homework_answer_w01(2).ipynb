Question 1: [Macro] Average growth of GDP in 2023

What is the average growth (in %) of GDP in 2023?

Download the timeseries Real Gross Domestic Product (GDPC1) from FRED (https://fred.stlouisfed.org/series/GDPC1). Calculate year-over-year (YoY) growth rate (that is, divide current value to one 4 quarters ago). Find the average YoY growth in 2023 (average from 4 YoY numbers). Round to 1 digit after the decimal point: e.g. if you get 5.66% growth => you should answer 5.7

```python
# install main library YFinance
!pip install yfinance
```

```python
# IMPORTS
import numpy as np
import pandas as pd

#Fin Data Sources
import yfinance as yf
import pandas_datareader as pdr
```

Answer:

```python
#because need for 4 YoY so the start from 2019
start = '2019-01-01'
end = '2023-12-31'
print(f'Period for indexes: {start} to {end} ')
```

```python
#Download data from (https://fred.stlouisfed.org/series/GDPC1)

gdpc1_data= pdr.DataReader('GDPC1', 'fred', start=start, end = end)
gdpc1_data.tail()
```

```python
# calculate Year Over Year (YoY) Growth rates
gdpc1_data['yoy_growth'] = gdpc1_data['GDPC1'].pct_change(periods=4)*100
gdpc1_data.tail()
```

```python
#calculate the average
average_yoy = gdpc1_data['yoy_growth'].mean()

print('Average YoY growth rate: ', (average_yoy))
```

Question 2. [Macro] Inverse "Treasury Yield"

Find the min value of (dgs10-dgs2) after since year 2000 (2000-01-01) and write it down as an answer, round to 1 digit after the decimal point.

Download DGS2 and DGS10 interest rates series (https://fred.stlouisfed.org/series/DGS2, https://fred.stlouisfed.org/series/DGS10). Join them together to one dataframe on date (you might need to read about pandas.DataFrame.join()), calculate the difference dgs10-dgs2 daily.

```python
# https://fred.stlouisfed.org/series/DGS2
start = '2000-01-01'
dgs2 = pdr.DataReader("DGS2", "fred", start=start)
dgs2.tail()
```

```python
# https://fred.stlouisfed.org/series/DGS10
dgs10 = pdr.DataReader("DGS10", "fred", start=start)
dgs10.tail()
```

```python
# join dgs2 and dgs10
dgs2_10 = dgs2.join(dgs10, lsuffix='_dgs2', rsuffix='_dgs10')
print(dgs2_10.tail())
```

```python
# calculate to find difference
dgs2_10['difference'] = dgs2_10['DGS10'] - dgs2_10['DGS2']
print(dgs2_10.tail())
```

```python
#find the minimum value
min_dgs2_10 = dgs2_10['difference'].min()
print("minimum value: ", round(min_dgs2_10, 1))
```

```python
# SPX = S&P500
# https://finance.yahoo.com/quote/%5EGSPC/

ticker_symbol = "^GSPC"
start = '2019-04-09'
end = '2024-04-09'

#download the data
spx_data = yf.download(ticker_symbol, start=start, end=end)
spx_data.tail()
```

```python
# IPC for Mexico
# https://finance.yahoo.com/quote/%5EMXX/

ticker_symbol = "^MXX"
start = '2019-04-09'
end = '2024-04-09'

#download the data
ipc_data = yf.download(ticker_symbol, start=start, end=end)
ipc_data.tail()
```

```python
ipc_data.head()
```

```python
# Calculate percentage change over the 5-year period
sp500_5y_growth = ((spx_data['Close'].iloc[-1] - spx_data['Close'].iloc[0]) / spx_data['Close'].iloc[0]) * 100
ipc_mexico_5y_growth = ((ipc_data['Close'].iloc[-1] - ipc_data['Close'].iloc[0]) / ipc_data['Close'].iloc[0]) * 100
```

Question 4. [Stocks OHLCV] 52-weeks range ratio (2023) for the selected stocks

Find the largest range ratio [=(max-min)/max] of Adj.Close prices in 2023

Download the 2023 daily OHLCV data on Yahoo Finance for top6 stocks on earnings (https://companiesmarketcap.com/most-profitable-companies/): 2222.SR,BRK-B, AAPL, MSFT, GOOG, JPM.

Here is the example data you should see in Pandas for "2222.SR": https://finance.yahoo.com/quote/2222.SR/history

Calculate maximum-minimim "Adj.Close" price for each stock and divide it by the maximum "Adj.Close" value. Round the result to two decimal places (e.g. 0.1575 will be 0.16)

(Additional: why this may be important for your research?)

```python
stock_symbols = ['2222.SR','BRK-B', 'AAPL', 'MSFT', 'GOOG', 'JPM']

# Initialize variables to store the largest range ratio and the corresponding stock symbol
largest_range_ratio = 0
largest_range_stock = None
```

```python
#download the data
for symbol in stock_symbols:
  stock_data = yf.download(symbol, start='2023-01-01', end='2023-12-31')
  # Calculate maximum and minimum adjusted close prices for the year
  max_close_price = stock_data['Adj Close'].max()
  min_close_price = stock_data['Adj Close'].min()

  # Calculate the range ratio
  range_ratio = (max_close_price - min_close_price) / max_close_price
  # Check if the current range ratio is the largest
  if range_ratio > largest_range_ratio:
      largest_range_ratio = range_ratio
      largest_range_stock = symbol
```

```python
# Print the largest range ratio and the corresponding stock symbol
print("Largest Range Ratio (2023):", round(largest_range_ratio, 2))
print("Stock Symbol:", largest_range_stock)
```

Question 5. [Stocks] Dividend Yield

Find the largest dividend yield for the same set of stocks

Use the same list of companies (2222.SR,BRK-B, AAPL, MSFT, GOOG, JPM) and download all dividends paid in 2023. You can use get_actions() method or .dividends field in yfinance library (https://github.com/ranaroussi/yfinance?tab=readme-ov-file#quick-start)

Sum up all dividends paid in 2023 per company and divide each value by the closing price (Adj.Close) at the last trading day of the year.

Find the maximum value in % and round to 1 digit after the decimal point. (E.g., if you obtained $1.25 dividends paid and the end year stock price is $100, the dividend yield is 1.25% -- and your answer should be equal to 1.3)

```python
# Initialize variables to store the largest dividend yield and the corresponding symbol
largest_dividend_yield = 0
largest_dividend_symbol = None

# Loop through each stock symbol
for symbol in stock_symbols:
    # Download dividends paid in 2023
    dividends = yf.Ticker(symbol).dividends
    # Sum up all dividends paid in 2023
    total_dividends = dividends.sum()

    # Get the closing price (Adj.Close) at the last trading day of the year
    stock_data = yf.download(symbol, end='2023-12-31')
    last_close_price = stock_data.iloc[-1]['Adj Close']
    # Calculate dividend yield
    dividend_yield = (total_dividends / last_close_price) * 100
```

```python
# Check if the current dividend yield is the largest seen so far
if dividend_yield > largest_dividend_yield:
  largest_dividend_yield = dividend_yield
  largest_dividend_symbol = symbol

# Print the largest dividend yield and the corresponding symbol
print("Largest Dividend Yield (2023):", round(largest_dividend_yield, 1), "%")
print("Symbol:", largest_dividend_symbol)
```
